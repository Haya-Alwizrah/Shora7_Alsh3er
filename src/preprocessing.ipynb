```python
import re
```

```python
def split_verses(text):
    pattern = r"\n?\d+\s*-\s*.*?(?=\n\d+\s*-\s*|$)"
    return re.findall(pattern, text, re.DOTALL)
```

```python
def parse_verse(block):
    # النص (بيت الشعر)
    text_part = block.split("المفردات.")[0].strip()

    vocab = re.search(r"المفردات\.(.*?)(المعنى)", block, re.DOTALL)
    meaning = re.search(r"المعنى(.*?)(الإعراب\.)", block, re.DOTALL)
    grammar = re.search(r"الإعراب\.(.*)", block, re.DOTALL)

    return {
        "text": text_part.strip(),
        "vocabulary": vocab.group(1).strip() if vocab else "",
        "meaning": meaning.group(1).strip() if meaning else "",
        "grammar": grammar.group(1).strip() if grammar else ""
    }
```

```python
def clean_text(text):
    text = re.sub(r"\n{2,}", "\n", text)
    return text.strip()
```

```python
def process_file(file_path):
    import os

    poet_name = os.path.basename(file_path).replace(".txt", "")

    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    text = clean_text(text)
    verses_blocks = split_verses(text)

    verses = [parse_verse(v) for v in verses_blocks]

    return {
        "poet": poet_name,
        "verses": verses
    }
```

```python
file_path = "/content/معلقة الأعشى.txt"
out = process_file(file_path)

out["verses"][0]
```
